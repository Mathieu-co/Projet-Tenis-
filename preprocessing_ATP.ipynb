```python
import glob
import pandas as pd
import numpy as np
filenames=list(glob.glob("Dataset/ATP/20*.xls*"))
l = [pd.read_excel(filename) for filename in filenames]
```

```python
no_b365=[i for i,d in enumerate(l) if "B365W" not in l[i].columns]
no_pi=[i for i,d in enumerate(l) if "PSW" not in l[i].columns]
for i in no_pi:
    l[i]["PSW"]=np.nan
    l[i]["PSL"]=np.nan
for i in no_b365:
    l[i]["B365W"]=np.nan
    l[i]["B365L"]=np.nan
l=[d[list(d.columns)[:13]+["Wsets","Lsets","Comment"]+["PSW","PSL","B365W","B365L"]] for d in [l[0]]+l[2:]]
data=pd.concat(l,0)
data.columns
```

```python
### Data cleaning
data=data.sort_values("Date")
data["WRank"]=data["WRank"].replace(np.nan,0)
data["WRank"]=data["WRank"].replace("NR",2000)
data["LRank"]=data["LRank"].replace(np.nan,0)
data["LRank"]=data["LRank"].replace("NR",2000)
data["WRank"]=data["WRank"].astype(int)
data["LRank"]=data["LRank"].astype(int)
data["Wsets"]=data["Wsets"].astype(float)
data["Lsets"]=data["Lsets"].replace("`1",1)
data["Lsets"]=data["Lsets"].astype(float)
data=data.reset_index(drop=True)
data
```

```python
# Series

# Define a mapping of tournament types to their respective integer values
tournament_mapping = {
    "ATP250": 250,
    "Grand Slam": 2000, 
    "Masters 1000": 1000,
    "ATP500": 500,
    "Masters Cup": 1500
}
# Create a new column with the integer values based on the mapping
data['tournament_value'] = data['Series'].map(tournament_mapping)
data['tournament_value'].value_counts()

```

```python
serie_round_drop = ["Series"]
data = data.drop(serie_round_drop,axis=1)

data["Series"] = data["tournament_value"]

data = data.drop(["tournament_value"],axis=1)

```

```python
data["Best of"] = pd.to_numeric(data['Best of'], errors='coerce')
```

```python
print(data.columns)
data.head()
```

```python
import random

from matplotlib import pyplot as plt

def lenght_verification(total,wanted, limite = 0.42):
    result = wanted
    pourcentage = (wanted/total)
    if pourcentage < limite:
        # print(f"erreur : {pourcentage}")
        result = limite*total
    elif pourcentage > 1-limite:
        # print(f"erreur : {pourcentage}")
        result = (1-limite)*total
    # else:
    #     print("GOOD")
    return int(result)

length_values = []
# Generate a random length following a Gaussian distribution
mean = len(data) / 2
std_deviation = 1000  # peux etre changé, plus il est grand plus la difference entre nbr 1 et nbr 0 est grandes

length = int(np.random.normal(mean, std_deviation))

length = lenght_verification(total=len(data),wanted=length)


y = [0] * length + [1] * (len(data) - length)
random.shuffle(y)

data['y'] = pd.Series(y, index=data.index)
data["y"].value_counts()
```

```python
# Divisez le DataFrame en deux en fonction de la colonne "y"
grouped = data.groupby('y')

# Accédez aux deux groupes
group_1_lose= grouped.get_group(0)
group_1_win = grouped.get_group(1)

lose_1 = group_1_lose.copy()
win_1 = group_1_win.copy()

lose_1.rename(columns={'Loser':'Player1_name',  
                       'LRank':'Player1_rank', 
                       'PSL':'Player1_PS',
                       'B365L':'Player1_B365',
                       
                       'Winner':'Player2_name',  
                       'WRank':'Player2_rank',
                       'PSW':'Player2_PS',
                       'B365W':'Player2_B365',},inplace=True)

win_1.rename(columns={'Loser':'Player2_name',  
                      'LRank':'Player2_rank',
                      'PSL':'Player2_PS',
                      'B365L':'Player2_B365',

                      'Winner':'Player1_name',  
                      'WRank':'Player1_rank', 
                      'PSW':'Player1_PS',
                      'B365W':'Player1_B365',},inplace=True)

all_matches_final= pd.concat([win_1,lose_1])
all_matches_final[["y","Player2_name","Player1_name"]]

all_matches_final = all_matches_final.reindex(range(len(all_matches_final)))
all_matches_final.head()
```

```python
corr_matrix = all_matches_final.corr()
corr_matrix['y'].sort_values(ascending=False)
```

# Multiclasse

```python
# On crée un nouveau datast
data.head(3)
data_multiclass = data.copy()
```

```python
# On crée une nouvelle varaible après avoir shuffle l'ancien dataset
def calculate_multiclass(row):
    win_sets = row['Wsets']
    lose_sets = row['Lsets']
    result = np.nan
    if win_sets == 2 and lose_sets == 0:
        result = '2 sets écart'
    elif win_sets == 2 and lose_sets == 1:
        result = '1 set écart'
    
    return result

data_multiclass['y_multicat'] = data_multiclass.apply(calculate_multiclass, axis=1)

data_multiclass = data_multiclass.dropna(subset=["y_multicat"])
```

```python
data_multiclass['y_multicat'].value_counts()
```

```python
grouped = data_multiclass.groupby('y')

# Accédez aux deux groupes
group_1_lose= grouped.get_group(0)
group_1_win = grouped.get_group(1)

lose_1 = group_1_lose.copy()
win_1 = group_1_win.copy()

lose_1["y_multicat"] = "Lose "+lose_1["y_multicat"]
win_1["y_multicat"] = "Win "+win_1["y_multicat"]
```

```python
lose_1["y_multicat"].value_counts()
```

```python
win_1["y_multicat"].value_counts()
```

```python
# On fusionne les deux dataset
final_dataset = pd.concat([lose_1, win_1], ignore_index=True)
final_dataset.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

final_dataset["y_multicat"] = label_encoder.fit_transform(final_dataset["y_multicat"])
final_dataset.head(2)
```

```python
# On enlève la variable y
all_matches_final_multi = final_dataset.drop('y', axis=1)
```

```python
# On enlève les NaN
all_matches_final_multi.dropna(subset=['B365W'], inplace=True)
all_matches_final_multi.dropna(subset=['B365L'], inplace=True)
all_matches_final_multi.dropna(subset=['PSW'], inplace=True)
all_matches_final_multi.dropna(subset=['PSL'], inplace=True)
```

```python
# Calcul de la médiane de la colonne
median_value = all_matches_final_multi['Best of'].median()

# Remplacement des NaN par la médiane
all_matches_final_multi['Best of'].fillna(median_value, inplace=True)
```
